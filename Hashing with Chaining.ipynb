# Introduction

Basically, a hash table is a data structure used to store key-value pairs, and it gives super-efficient insertion, search, and deletion operations. A hash function maps each key to a specific index in an underlying array, which allows us to directly access the stored elements. But since different keys can end up producing the exact same hash value, collisions are bound to happen. One of the most common ways to resolve this is chaining, where each slot in the hash table simply maintains a linked list (or chain) of all the key-value pairs that hash to that same index.

In this implementation, I have used chaining to handle collisions efficiently. To distribute the keys uniformly across the table, a simple universal-style hash function based on Python’s built-in hash() function is used. The code supports three fundamental operations: inserting new key-value pairs, searching for an existing key, and deleting a key-value pair. On top of that, the implementation dynamically resizes the table whenever the load factor goes too high, which heavily helps in maintaining optimal performance without any lag.

Hash Table Implementation Using Chaining

```python
class HashTable:
    """
    Hash Table implementation using chaining for collision resolution.
    Each bucket stores a list of (key, value) pairs.
    """

    def __init__(self, capacity=10):
        """
        Initialize the hash table.

        Parameters:
            capacity (int): Number of buckets in the table.
        """
        self.capacity = capacity
        self.size = 0

        # Create empty buckets
        self.table = [[] for _ in range(capacity)]

    def hash_function(self, key):
        """
        Compute the bucket index for a given key.

        Python's built-in hash function provides good distribution,
        and the modulo operation maps the value to a valid bucket index.
        """
        return hash(key) % self.capacity

    def insert(self, key, value):
        """
        Insert or update a key-value pair.

        If the key already exists, its value is updated.
        Otherwise, a new key-value pair is added.
        """

        index = self.hash_function(key)
        bucket = self.table[index]

        # Update existing key
        for i, (k, v) in enumerate(bucket):
            if k == key:
                bucket[i] = (key, value)
                return

        # Insert new key-value pair
        bucket.append((key, value))
        self.size += 1

        # Resize if load factor exceeds 0.75
        if self.load_factor() > 0.75:
            self.resize()

    def search(self, key):
        """
        Search for a key.

        Returns:
            Associated value if found,
            otherwise None.
        """

        index = self.hash_function(key)

        bucket = self.table[index]

        for k, v in bucket:
            if k == key:
                return v

        return None

    def delete(self, key):
        """
        Remove a key-value pair from the table.

        Returns:
            True if deleted,
            False if the key does not exist.
        """

        index = self.hash_function(key)

        bucket = self.table[index]

        for i, (k, v) in enumerate(bucket):

            if k == key:
                del bucket[i]
                self.size -= 1
                return True

        return False

    def load_factor(self):
        """
        Calculate the current load factor.
        """
        return self.size / self.capacity

    def resize(self):
        """
        Double the size of the hash table and
        rehash all existing elements.
        """

        old_table = self.table

        self.capacity *= 2

        self.table = [[] for _ in range(self.capacity)]

        self.size = 0

        for bucket in old_table:
            for key, value in bucket:
                self.insert(key, value)
```

Testing the Hash Table

The following examples verify that the hash table correctly supports insertion, searching, updating, and deletion of key-value pairs.

```python
hash_table = HashTable()

# Insert key-value pairs
hash_table.insert("Alice", 91)
hash_table.insert("Bob", 85)
hash_table.insert("Charlie", 78)
hash_table.insert("David", 95)

# Search operations
print("Alice:", hash_table.search("Alice"))
print("Charlie:", hash_table.search("Charlie"))

# Update an existing key
hash_table.insert("Alice", 99)

print("Updated Alice:", hash_table.search("Alice"))

# Delete a key
hash_table.delete("Bob")

print("Deletion search for Bob:", hash_table.search("Bob"))

print("Current Load Factor:", hash_table.load_factor())
```
